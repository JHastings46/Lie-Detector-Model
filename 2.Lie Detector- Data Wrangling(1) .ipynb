#### Feature Engineering 

```python
# Compute sum of scores for each row
row_total = thp_df[['PWS','DL','SC','RC','HMM','WSL','CEP','LF']].sum(axis=1)

# Calculate the grand total of all scores
grand_total = row_total.sum()

# Normalize each individual's total score by the grand total
thp_df['fraud_alert'] = row_total / grand_total

#incorperate varability by adding another column that calculates std across rows
thp_df['sigma'] = thp_df[['PWS','DL','SC','RC','HMM','WSL','CEP','LF']].std(axis=1)
```

```python
#normalize the data with z-score(fraud_alert normalized)
thp_df[['PWS', 'DL', 'SC', 'RC', 'HMM', 'WSL', 'CEP', 'LF','sigma','fraud_alert']] = (
    thp_df[['PWS', 'DL', 'SC', 'RC', 'HMM', 'WSL', 'CEP', 'LF','sigma','fraud_alert']] - 
    thp_df[['PWS', 'DL', 'SC', 'RC', 'HMM', 'WSL', 'CEP', 'LF', 'sigma','fraud_alert']].mean()
) / thp_df[['PWS', 'DL', 'SC', 'RC', 'HMM', 'WSL', 'CEP', 'LF','sigma','fraud_alert']].std()
thp_df.head()
```

```python
#fraud_alert not normalized
#normalize the data with z-score
thp_df[['PWS', 'DL', 'SC', 'RC', 'HMM', 'WSL', 'CEP', 'LF','sigma']] = (
    thp_df[['PWS', 'DL', 'SC', 'RC', 'HMM', 'WSL', 'CEP', 'LF','sigma',]] - 
    thp_df[['PWS', 'DL', 'SC', 'RC', 'HMM', 'WSL', 'CEP', 'LF', 'sigma',]].mean()
) / thp_df[['PWS', 'DL', 'SC', 'RC', 'HMM', 'WSL', 'CEP', 'LF','sigma',]].std()
thp_df.head()
```

```python
#Removing Outliers for 'new_clean.csv' version
# Calculate Q1, Q3 and IQR for each column
Q1 = thp_df.quantile(0.25)
Q3 = thp_df.quantile(0.75)
IQR = Q3 - Q1

# Define a mask for values that are not outliers
mask = ~((thp_df < (Q1 - 1.5 * IQR)) | (thp_df > (Q3 + 1.5 * IQR))).any(axis=1)

# Apply the mask to get a DataFrame without outliers
remove_outliers = thp_df[mask]

# Display the shape of the DataFrame without outliers
print(remove_outliers.shape)
```

```python
#Adding new column & then rearrange columns

thp_df['total_score']=thp_df[["PWS","DL","SC","RC","HMM","WSL","CEP","LF"]].sum(axis=1)

# Get a list of all column names
cols = list(thp_df.columns)

# Decide the position where you want your new column (next to the column 'LF' in this case)
# Get the index of the column after which you want to place the new column
position = cols.index('LF') + 1

# Reorder the columns
# This will slice all columns up to the position, add the 'total_score' column, and then add the rest
cols = cols[:position] + ['total_score'] + cols[position:-1]

# Reassign the DataFrame columns to the new order
thp_df = thp_df[cols]
```
